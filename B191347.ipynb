```python
rooms_count=df["Rooms"].value_counts()
```

```python
rooms_count=dict(rooms_count)
```

```python
rooms_count
```

```python
a=rooms_count.keys()
b=rooms_count.values()
fig=plt.figure()
plt.title("Piechart")

plt.pie(x=b,labels=a)
plt.legend()
plt.show()
```

```python
region=df['Regionname']
region_count=df['Regionname'].value_counts()
region_count=dict(region_count)
a=region_count.keys()
b=region_count.values()
fig=plt.figure(figsize=(7,7))
plt.title("Piechart")

plt.pie(x=b,labels=a)
plt.legend()
plt.show()
```

```python
type_count=df["Type"].value_counts()
type_count=dict(type_count)
print(type_count)
a=type_count.keys()
b=type_count.values()
```

```python
#fig=plt.fig()
plt.bar(a,b)
```

```python
suburb_count=df["Suburb"].value_counts()
suburb_count=dict(suburb_count)
print(suburb_count)
a=suburb_count.keys()
b=suburb_count.values()
```

```python
fig=plt.figure(figsize=(100,100))
plt.bar(b,a)
plt.show()
```

```python
m=df["Method"].value_counts()
m=dict(m)
print(m)
a=m.keys()
b=m.values()
fig=plt.figure(figsize=(4,4))
plt.bar(x=a,height=b)
plt.show()
```

```python
m=df["Method"].value_counts()
m=dict(m)
print(m)
a=m.keys()
b=m.values()
fig=plt.figure(figsize=(4,4))
plt.fill_between(a,b)
plt.show()
```
